```python
import pandas as pd
import folium
from matplotlib.colors import Normalize, rgb2hex
import matplotlib.cm as cm
```

### Get the latest Earthquake data from USGS.  Compare to:
http://earthquake.usgs.gov/earthquakes/map/
    
Note:  This Jupyter notebook does not update itself.  You will have to download and run it to get a true real-time earthquake map.

```python
data = pd.read_csv('http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.csv')
data.head()
```

```python
non_quakes = []
for i,j in enumerate(data.type):
    if j != "earthquake":
        non_quakes.append(i)
        print i
    else:
        pass
if len(non_quakes) == 0:
    print "All seismic activity recorded are earthquakes."
else:
    print "Seismic activity non-earthquakes are itmes: %r"%(non_quakes)
    pass
```

```python
# normalize the colors used based on max and min magnitudes of current earthquakes.
norm = Normalize(data['mag'].min(), data['mag'].max())
```

```python
map = folium.Map(location=[48, -102], zoom_start=3)#, crs='EPSG4326')
for eq in data.iterrows():
    color = rgb2hex(cm.OrRd(norm(float(eq[1]['mag']))))
    info = eq[1]['place']+' at '+eq[1]['time']+'. Magnitude: '+str(eq[1]['mag'])
    map.circle_marker([eq[1]['latitude'], eq[1]['longitude']], 
                    popup= info, 
                    radius=20000*float(eq[1]['mag']),
                    line_color=color,
                    fill_color=color)
map.save(outfile='earthquake.html')
```

# A map of my favorite Mountian Bike Trails in NW Arkansas.
This map is very basic at the moment and I will add to it with time.

```python
from folium import features
```

```python
map_1 = folium.Map(location=[36.2, -94.16], zoom_start=9,
                   tiles='Stamen Terrain')

bc = features.Icon(color='blue')
rc = features.Icon(color='red')
kc = features.Icon(color='black')

Ta = features.Marker([36.038903, -94.221840], popup='Kessler Mountain Trails')
Tb = features.Marker([36.379506, -94.205891], popup='Slaughter Penn Trails')
Tc = features.Marker([36.291820, -93.976314], popup='Hobbs State Park Trails')
Td = features.Marker([36.435329, -93.760566], popup='Lake Leatherwood Trails')

map_1.add_children(Ta)
map_1.add_children(Tb)
map_1.add_children(Tc)
map_1.add_children(Td)
map_1.save(outfile='ARtrails2.html')
```
